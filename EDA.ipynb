```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

## PASO 2: Obtener los datos histÃ³ricos de las empresas desde el 01/01/1990:

```python
start_date = pd.datetime.now().date() - pd.DateOffset(years=33)
end_date = pd.datetime.now().strftime('%Y-%m-%d')
end_date = pd.to_datetime(end_date).date()  # convertir a datetime.date
companies = ['AOS', 'AFL', 'ABT', 'PTC']

data = pd.DataFrame()
for company in companies:
    df = yf.download(company, start=start_date, end=end_date)
    df['Company'] = company
    data = pd.concat([data, df])
```

## PASO 3: Visualizar el conjunto de datos y explorar sus estadÃ­sticas descriptivas:

```python
print(data.head())
print(data.describe())
```

```python
data['Date'] = data.index.date
data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]
```

## PASO 4: Visualizar la distribuciÃ³n de las variables numÃ©ricas:

date debe ser x y Close y

```python
sns.lineplot(data=data, x='Date', y='Close', hue='Company')
plt.xlabel('Fechas histÃ³ricas de los Ãºltimos 33 aÃ±os')
plt.ylabel('Precio de cierre')
plt.title('DistribuciÃ³n del precio de cierre por empresa')
plt.show()
```

## PASO 5: Realizar un anÃ¡lisis bivariado de las variables numÃ©ricas y/o categÃ³ricas:

```python
sns.pairplot(data=data, hue='Company')
plt.show()
```

## PASO 6: Identificar patrones, outliers y/o anomalÃ­as en los datos:

```python
sns.boxplot(data=data, x='Company', y='Close')
plt.xlabel('Empresa')
plt.ylabel('Precio de cierre')
plt.title('Boxplot del precio de cierre por empresa')
plt.show()

sns.boxplot(data=data, x='Company', y='Volume')
plt.xlabel('Empresa')
plt.ylabel('Volumen')
plt.title('Boxplot del volumen por empresa')
plt.show()
```
